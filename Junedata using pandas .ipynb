```python
import pandas as pd
sales = pd.read_csv("junesales.csv", header = 0)
sales.head()
```

```python
sales["daysofmonth"] = pd.DatetimeIndex(sales["Date"]).day
sales.head()
```

```python
data = pd.read_csv("weatherdata.csv",header = 0)
data.head()
```

```python
data["daysofmonth"] = pd.DatetimeIndex(data["Date"]).day
data["Year"] = pd.DatetimeIndex(data["Date"]).year
data["Month"] = pd.DatetimeIndex(data["Date"]).month
data.head()
```

```python
Newcastle_data = data[(data["Location"]== "Newcastle") & (data["Year"] ==2011) & (data["Month"]== 6)]
Newcastle_data.head()
```

```python
merge_data = Newcastle_data.merge(sales, on = "daysofmonth")
merge_data.head()
```
