## 1.6 Grouping in Pandas
<hr/>

```python
fortune = pd.read_csv('fortune1000.csv',index_col='Rank')
```

```python
fortune.head()
```

```python
secGrp = fortune.groupby('Sector')
```

```python
# POINT TO NOTE

There are 21 Groups(DataFrames) inside secGrp Object.

Each Group is Named After Sector Name

you can fetch each group using get_group(<sectorName>)
```

```python
secGrp.get_group('Health Care')['Profits'].sum()
```

```python
secGrp.mean()
```

```python
# .agg()

{'column':'operation'}
```

```python
secGrp.agg({'Revenue':'sum','Profits':['sum','mean'],'Employees':['max','min','sum','mean']})
```
