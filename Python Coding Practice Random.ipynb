```python
import random
```

```python
jackpot = random.randint(1,100)
guess = int(input("Please guess the number: ")) 
counter = 1
while guess != jackpot: 
    if guess < jackpot: 
        print("Guess Higher")
    else: 
        print("Gues Lower") 

    guess = int(input("Chal guess kar")) 
    counter += 1

print("Right Guess!") 
print("You took", counter, "attempts")
```
