```python
# finding the duplicates in the given list 
def find_duplicates_fast(lst):
    seen = set()
    duplicates = []
    dup_set = set()            # keep track of duplicates we already recorded
    for item in lst:
        if item in seen and item not in dup_set:
            duplicates.append(item)
            dup_set.add(item)
        else:
            seen.add(item)
    return duplicates
```

```python
numbers = [1, 2, 3, 4, 2, 3, 5, 6, 1]
print(find_duplicates(numbers))
```

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr
```

```python
numbers_list = [4,10,5,15,18,7,8]
print(bubble_sort(numbers_list))
```

```python
# Check if a number is prime
num = int(input("Enter a number: "))

if num > 1:
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            print(f"{num} is not a prime number.")
            break
    else:
        print(f"{num} is a prime number.")
else:
    print(f"{num} is not a prime number.")
```

```python
# to check if the number is factorial 
def factorial(num): 
    if num < 0: 
        return "factorial number doesn't exist for negative numbers"
    elif num == 0 and num == 1: 
        return 1 
    else: 
        fact = 1 
        for i in range (2, num + 1): 
            fact *= i 
        return fact 
print(factorial(5))
```

```python
def fibonacci(n):
    if n <= 0:
        return "Invalid input"
    elif n == 1:
        return 0
    elif n == 2:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)

# Example usage
print(fibonacci(10)) 
```

```python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
           # map each character to its last seen index + 1 (to make left jumps easy)
        charSet = set() 
        l = 0 
        res = 0 

        for r in range(len(s)): 
            while s[r] in charSet: 
                charSet.remove(s[l]) 
                l += 1 
            charSet.add(s[r]) 
            res = max(res, r - l + 1) 
        return res
```

```python
def first_non_repeated_char(s):
    
    char_count = {}
    
    for char in s:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1
   
    for char in s:
        if char_count[char] == 1:
            return char
    
    return None


print(first_non_repeated_char("swiss")) 
print(first_non_repeated_char("aabbcc"))
```
